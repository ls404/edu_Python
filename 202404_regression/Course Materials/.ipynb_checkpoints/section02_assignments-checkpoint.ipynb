## Assignment 2: Simple Regression

```python
computers.plot.scatter(x="ram", y="price").set_ylim(0, 10000)
```

```python
import statsmodels.api as sm

X = sm.add_constant(computers["ram"])
y = computers["price"]

model = sm.OLS(y, X).fit()

model.summary()
```

```python
# interpret the above
```

```python
ram_values = [0, 2, 4, 8, 16, 32, 64]

pred = model.predict(sm.add_constant(pd.DataFrame({"ram": ram_values})))

pd.DataFrame(zip(ram_values, pred)).rename({0: "ram", 1: "predicted_price"}, axis=1)
```
