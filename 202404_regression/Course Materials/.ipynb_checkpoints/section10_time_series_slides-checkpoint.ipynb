### Time Series Behavior

```python
import pandas as pd
import numpy as np
import seaborn as sns

data = pd.read_csv(
    "train.csv", 
    parse_dates=["Order Date", "Ship Date"], 
    dayfirst=False,
) # ignore warning
```

```python
# Daily aggregation

ts_df = data.sort_values(by="Order Date").groupby("Order Date").agg({"Sales": "sum"})

ts_df.head()
```

```python
# Monthly Aggregation

ts_df.resample("M").sum().head()
```

```python
# Yearly Aggregation

ts_df.resample("Y").sum().head()
```

```python
ts_df.resample("M").sum().plot()

sns.despine()
```

```python
monthly_sales = (
    ts_df
    .resample("M")
    .sum()
    .assign(
        Quarterly_Window=lambda x: x["Sales"].rolling(window=3).mean(),
        half_year=lambda x: x["Sales"].rolling(window=6).mean(),
        Annual_Window=lambda x: x["Sales"].rolling(window=12).mean()
    )
)

monthly_sales.head()
```

```python
monthly_sales[["Sales", "Quarterly_Window"]].plot(color=["grey", "red"])

sns.despine()
```

```python
monthly_sales[["Sales", "Annual_Window"]].plot(color=["grey", "red"])

sns.despine()
```

```python
exponential_smoothing_df = (
    ts_df
    .resample("M")
    .sum()
    .assign(
        EWM_1=lambda x: x["Sales"].ewm(alpha=.1).mean(),
        EWM_5=lambda x: x["Sales"].ewm(alpha=.5).mean()
    )
)

exponential_smoothing_df.head()
```

```python
exponential_smoothing_df.plot(color=["grey", "green", "red"])

sns.despine()
```

### Additive Vs. Multiplicative Series

```python
# Additive data creation

import numpy as np

seasonal_adders = [11, 13, 12, 15, 19, 23, 21, 28, 20, 17, 15, 12]
# Immitate 10 years of data
x_range = np.arange(1, 121)

time_series = []
# Split to 10 chunks - 1 year each
for chunk in np.split(x_range, 10):
    for i, val in enumerate(chunk):
        # Add value to seasonal scalar
        time_series.append(val + seasonal_adders[i])

# Convert to DataFrame
df_aa = pd.DataFrame(
    data={'Taco Stands in LA': time_series}, 
    index=pd.date_range(start='2010-01-01', freq='MS', periods=120)
)

# Plot
df_aa.plot(ylabel="Taco Stands in LA");

sns.despine()
```

### Holt Winters Forecasting

```python
weather = pd.read_csv(
    "madrid_weather.csv", 
    usecols=["temperature", "time"], 
    parse_dates=True,
    index_col="time"
)

weather.head()
```

```python
weather_train = weather.loc["2019-01-01": "2019-05-31"]
weather_test = weather.loc["2019-06-01": "2019-06-03"]
```

```python
weather#.loc["2019-01-01"]
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

smoothing_model = (
    ExponentialSmoothing(
        weather_train, 
        trend="additive", 
        seasonal="additive", 
        seasonal_periods=24
    ).fit()
)
```

```python
smoothing_model.summary()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

weather_test.plot()
smoothing_model.forecast(72).plot()


ax.legend(["Actual Values", "Predicted Values"])

ax.spines[["top", "right"]].set_visible(False);
```

```python
from sklearn.metrics import mean_absolute_percentage_error as mape
from sklearn.metrics import mean_absolute_error as  mae

weather_test = weather_test.assign(predictions = smoothing_model.forecast(72))


print(
    f"MAE: {mae(weather_test['temperature'], weather_test['predictions'])}",
    f"MAPE: {mape(weather_test['temperature'], weather_test['predictions'])}",
)
```

### Regression With Trend and Season

```python
weather = pd.read_csv(
    "madrid_weather.csv", 
    usecols=["temperature", "time"], 
    parse_dates=["time"],
#     index_col="time"
)

weather.head()
```

```python
weather = pd.get_dummies(
    weather.assign(
    trend = weather.index + 1,
    hour = weather["time"].dt.hour.astype("string")
    ), 
    drop_first=True
).set_index("time")
```

```python
weather.head()
```

```python
weather_train = weather["2019-05-17": "2019-05-31"]
weather_test = weather["2019-06-01": "2019-06-03"]
```

```python
import statsmodels.api as sm

y = weather_train["temperature"]
X = sm.add_constant(weather_train.iloc[:, 1:])

model = sm.OLS(y, X).fit()

model.summary()
```

```python
weather_test = (
    weather_test
     .assign(predictions = model.predict(sm.add_constant(weather_test.iloc[:, 1:])))
     .rename({"temperature": "actual"}, axis=1)
     .loc[:, ["actual", "predictions"]]
)
 
weather_test.plot(
     title="Predicted Vs. Actual Weather - Test Set",
     ylabel="Temperature (C)",
)

sns.despine()
```

```python
from sklearn.metrics import mean_absolute_percentage_error as mape
from sklearn.metrics import mean_absolute_error as  mae

print(
    f"MAE: {mae(weather_test['actual'], weather_test['predictions'])}",
    f"MAPE: {mape(weather_test['actual'], weather_test['predictions'])}",
)
```

### Facebook Prophet

```python
weather = pd.read_csv(
    "madrid_weather.csv", 
    usecols=["temperature", "time"], 
    parse_dates=["time"],
    index_col="time"
)

weather.head()
```

```python
weather_train = weather["2019-05-17": "2019-05-31"].reset_index()
weather_test = weather["2019-06-01": "2019-06-03"].reset_index()
```

```python
# !conda install prophet --y

from prophet import Prophet

weather_train = weather_train.rename({"time": "ds", "temperature": "y"}, axis=1)

weather_train.head()
```

```python
model = Prophet()
model.fit(weather_train)
```

```python
future = model.make_future_dataframe(periods=72, freq="H")

forecast = model.predict(future)

model.plot(forecast)
```

```python
model.plot_components(forecast)
```

```python
weather_test = weather_test.assign(predictions = forecast.set_index("ds").loc["2019-06-01":"2019-06-03", "yhat"].values)
```

```python
weather_test.set_index("time").plot(ylim=0, ylabel="Temperature (C)")
```

```python
from sklearn.metrics import mean_absolute_percentage_error as mape
from sklearn.metrics import mean_absolute_error as  mae

print(
    f"MAE: {mae(weather_test['temperature'], weather_test['predictions'])}",
    f"MAPE: {mape(weather_test['temperature'], weather_test['predictions'])}",
)
```
