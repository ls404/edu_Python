```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm

insurance_df = pd.read_csv("/Data/insurance.csv")
```

```python
insurance_df.head()
```

```python
features = ["age", "bmi"]

X = sm.add_constant(insurance_df[features])
y = insurance_df["price"]

model = sm.OLS(y, X).fit()

model.summary()
```

```python
new_data = pd.DataFrame({
    "age": [0, insurance_df["age"].min(), insurance_df["age"].mean(), insurance_df["age"].max()],
    "bmi": [0, insurance_df["bmi"].min(), insurance_df["bmi"].mean(), insurance_df["bmi"].max()],
})

new_data
```

```python
X = sm.add_constant(new_data)

model.predict(X)
```

```python
insurance_df = insurance_df.assign(
    smoker_flag=np.where(insurance_df["smoker"] == "yes", 1, 0),
    female_flag=np.where(insurance_df["sex"] == "female", 1, 0)
)

X = sm.add_constant(insurance_df[["age", "bmi", "children", "smoker_flag"]])
y = insurance_df["price"]

model = sm.OLS(y, X).fit()

model.summary()
```

```python
sns.scatterplot(x=model.predict(), y=model.resid).set(ylim=(-13000, 50000) )
```

```python
sns.scatterplot(x=insurance_df["age"], y=insurance_df["price"])
```

```python
import numpy as np

rng = np.random.default_rng(12345)

insurance_df = insurance_df.assign(
    rand_col = rng.random(len(insurance_df)),
    rand_col2 = rng.random(len(insurance_df)),
    rand_col3 = rng.random(len(insurance_df)),
)
```

```python

cols = ["age", "bmi", "children", #"smoker_flag", 
        "rand_col", "rand_col2", "rand_col3", 
       ]

X = sm.add_constant(insurance_df[cols])
y = insurance_df["price"]

model = sm.OLS(y, X).fit()

model.summary()
```

```python
insurance_df = insurance_df.merge(
    pd.get_dummies(insurance_df["region"]),
    left_index=True,
    right_index=True
)

features = ["age", "age_sq", "bmi", "children", "smoker_flag", "northwest", "southeast", "southwest"]


X = sm.add_constant(insurance_df[features])
y = insurance_df["price"]


model = sm.OLS(y, X).fit()

model.summary()
```
