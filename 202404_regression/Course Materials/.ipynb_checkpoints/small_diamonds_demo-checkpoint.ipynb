```python
# Function to plot actual X vs. Y's along with regression prediction

def model_fit_vs_x(x, y):
    import statsmodels.api as sm
    import matplotlib.pyplot as plt
    model = sm.OLS(y, sm.add_constant(x)).fit()
    pred = model.predict(sm.add_constant(x))
    fig, ax = plt.subplots()
    ax.scatter(x, y)
    ax.plot(x, pred, color="blue", ls="--")
    ax.set_title("Regression Fit")
    ax.set_xlabel(x.name)
    ax.set_ylabel(y.name)
    

    
model_fit_vs_x(diamonds["carat"], diamonds["price"])
    
```

# Model Assumptions

In terms of assumptions:

1. Linarity: We still haven't addressed potential non-linear relationships between features and target based on pairplot scatterplots
2. Independence of Errors: Durbin-Watson Statistic near 2, so this is good!
3. Normality: JB-Test indicates residuals are not normally distributed
4. No-Perfect-Multi-Collinearity: Our Condition Number Warning fired indicating collinearity issues. Our variable signs also don't match scatterplot relationships for a few variables (e.g. `x`).
5. Equal Variance of Errors. Variance is very odd,Almost constant if we rotate residual cloud 45 degrees. We'll assess this after addressing others later. Our high-carat diamonds may still be causing issues with high predictions

```python
# residual Analysis Plotting function - residual plot and QQplot

def residual_analysis_plots(model):
    
    import scipy.stats as stats
    import matplotlib.pyplot as plt
    
    predictions = model.predict()
    residuals = model.resid
    
    fig, ax = plt.subplots(1, 2, sharey="all", figsize=(10, 6))
    
    sns.scatterplot(x=predictions, y=residuals, ax=ax[0])
    ax[0].set_title("Residual Plot")
    ax[0].set_xlabel("Prediction")
    ax[0].set_ylabel("Residuals")
    
    stats.probplot(residuals, dist="norm", plot=ax[1])
    ax[1].set_title("Normal Q-Q Plot")   
```

```python
# Residuals Look odd, but constant variance
# QQ-plot is close to normal between -2, 2 standard deviations but a bit skewed

residual_analysis_plots(model)    
```

Let's tackle our violated assumptions. First, let's tackle the collinear variables. We could tackle these issues in a number of ways, but solidifying our final variable set will prevent backtracking later. 

To do this, we'll look at variable VIF to identify the potential problem variables. 

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif

pd.Series(
    [vif(X.values, i) for i in range(X.shape[1])],
    index=X.columns
)
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score as r2


kf = KFold(n_splits=5, shuffle=True, random_state=2023)

# Create a list to store validation scores for each fold
cv_lm_r2s = []
cv_lm_mae = []

# Loop through each fold in X and y
for train_ind, val_ind in kf.split(X, y):
    # Subset data based on CV folds
    X_train, y_train = X.iloc[train_ind], y.iloc[train_ind]
    X_val, y_val = X.iloc[val_ind], y.iloc[val_ind]
    # Fit the Model on fold's training data
    model = sm.OLS(y_train, X_train).fit()
    # Append Validation score to list 
    cv_lm_r2s.append(r2(y_val, model.predict(X_val),))
    cv_lm_mae.append(mae(y_val, model.predict(X_val),))

print("All Validation R2s: ", [round(x, 3) for x in cv_lm_r2s])
print(f"Cross Val R2s: {round(np.mean(cv_lm_r2s), 3)} +- {round(np.std(cv_lm_r2s), 3)}")

print("All Validation MAEs: ", [round(x, 3) for x in cv_lm_mae])
print(f"Cross Val MAEs: {round(np.mean(cv_lm_mae), 3)} +- {round(np.std(cv_lm_mae), 3)}")
```

### Polynomial Terms

* We've seen an example of Capturing Non-linear relationships with a squared term. 
* Let's take a look at a cubed term, 4th and 5th power terms. 
* You can add any degree polynomial relationship, however it's very rare to go beyond a cubed term, and its uncommon to go past squared. As these terms get more complex, the risk of overfitting grows.
* The cubed term appears to help, but the 4th and 5th power terms start to reduce out of sample performance

```python
diamonds = diamonds.assign(
    carat_cubed = diamonds["carat"] ** 3,
    carat_4th = diamonds["carat"] ** 4,
    carat_5th = diamonds["carat"] ** 5
)

cols = [
    "carat", 
    "carat_sq", 
    "carat_cubed",
    "carat_4th",
    "carat_5th",
    "depth", 
    "table", 
#     "x", 
#     "y",
#     "z"
]

X = sm.add_constant(diamonds.loc[:, cols])
# y = diamonds["price"]
y = np.log(diamonds["price"])


# Create a list to store validation scores for each fold


def cross_val(X, y, n_splits):
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=2023)
    # Create a list to store validation scores for each fold

    cv_lm_r2s = []
    cv_lm_mae = []
    
    for train_ind, val_ind in kf.split(X, y):
        # Subset data based on CV folds
        X_train, y_train = X.iloc[train_ind], y.iloc[train_ind]
        X_val, y_val = X.iloc[val_ind], y.iloc[val_ind]
        # Fit the Model on fold's training data
        model = sm.OLS(y_train, X_train).fit()
        # Append Validation score to list 
        cv_lm_r2s.append(r2(y_val, model.predict(X_val),))
        cv_lm_mae.append(mae(y_val, model.predict(X_val),))

    print("All Validation R2s: ", [round(x, 3) for x in cv_lm_r2s])
    print(f"Cross Val R2s: {round(np.mean(cv_lm_r2s), 3)} +- {round(np.std(cv_lm_r2s), 3)}")

    print("All Validation MAEs: ", [round(x, 3) for x in cv_lm_mae])
    print(f"Cross Val MAEs: {round(np.mean(cv_lm_mae), 3)} +- {round(np.std(cv_lm_mae), 3)}")
    
cross_val(X, y, 5)

print(f"Test R2: {r2(model.predict(X_test), y_test)}")
print(f"Test MAE: {mae(model.predict(X_test), y_test)}")      
print(f"Test MSE: {mse(model.predict(X_test), y_test) ** 1/2}")    
```

### One-Hot Encoding/Dummy Variables

Regression models (and most ML algos) require numeric inputs. We can convert categorical data into numeric data with the help of one hot encoding, which represents each category as a binary variable.

For example, if you had a column with two possible values, "Yes", and "No", "Yes" can be represented by 1, and "No" by 0. The coefficient assigned to this variable essentially changes the intercept of our model by the magnitude of the coefficient if the value is 1. If 0, our predictions don't change. 

We'll show a few examples of this.

1. A simple Binary Column to represent two possibilites
2. Dummy variables (one hot encoding) to represent multiple categories
3. The Dummy Variable Trap
4. Ordinal Encoding

### Dummy Variables

* We will use pandas get_dummies method to one-hot encode our categorical variables.
* I suggest looking at the value counts method before doing this. If there are rare categories < ~10 values, we should consider folding them into other categories as our data splitting may mean our holdout data doesn't contain observations from these values. 
* Plotting bar charts of your categorical variable vs. target can inform you whether there is a significant difference between categories. If no significant difference, we can likely move on. IF only one category pops, we can do 1 vs. all in a single binary column.

```python
def value_counter(dataframe):
    value_series = pd.DataFrame()
    for col in dataframe.select_dtypes(["object"]).columns:
        print(dataframe[col].value_counts())
        
        
value_counter(diamonds)
```

```python
def cat_plotter(data, target):
    for col in data.select_dtypes(["object"]).columns:
        sns.barplot(data=data, x=col, y=target)
        plt.xticks(rotation=45)
        plt.show()
        
cat_plotter(diamonds, "price")
```

### Binning Data

* We can also reduce the number of categories in a column by combining, or binning them. This can help reduce overfitting when we have many categories, and some categories are rare. It can also help simplify model interpretation.

```python
from sklearn.linear_model import LinearRegression, Lasso, LassoCV, Ridge, RidgeCV
# from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

lr_model_ridge = Ridge(alpha = 100)
lr_model_ridge.fit(X_tr, y_train)
ridge_test_r2 = lr_model_ridge.score(X_te, y_test)

print(lr_model_ridge.score(X_tr, y_train))
print(lr_model_ridge.score(X_te, y_test))

list(zip(X_train.columns, lr_model_ridge.coef_))
```

```python
coefs = []
errors = []

alphas = np.logspace(-3, 5, 200)

model = Ridge()

# Train the model with different regularisation strengths
for a in alphas:
    model.set_params(alpha=a)
    model.fit(X_tr, y_train)
    coefs.append(model.coef_)
    errors.append(mae(y_test, model.predict(X_te)))

# Display results
plt.figure(figsize=(20, 6))

plt.subplot(121)
ax = plt.gca()
ax.plot(alphas, coefs)
ax.set_xscale("log")
plt.xlabel("alpha")
plt.ylabel("weights")
plt.title("Ridge coefficients as a function of the regularization")
plt.axis("tight")

plt.subplot(122)
ax = plt.gca()
ax.plot(alphas, errors)
ax.set_xscale("log")
plt.xlabel("alpha")
plt.ylabel("error")
plt.title("Coefficient error as a function of the regularization")
plt.axis("tight")

plt.show()
```
