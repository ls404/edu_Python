```python
import pandas as pd
import numpy as np

import statsmodels.api as sm
import seaborn as sns

anscombe = pd.read_csv("anscombe.csv").drop("Unnamed: 0", axis=1)

anscombe.head()
```

```python
anscombe["Product"].value_counts()
```

```python
coffee = anscombe.query("Product == 'Coffee'")
```

```python
sns.lmplot(coffee, x="Units Sold", y="Profit", ci=None)
```

```python
tea = anscombe.query("Product == 'Tea'")

sns.lmplot(tea, x="Units Sold", y="Profit", ci=None)
```
