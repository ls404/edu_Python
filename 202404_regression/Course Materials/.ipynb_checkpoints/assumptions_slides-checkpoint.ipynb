```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm

diamonds = pd.read_csv("Diamonds Prices2022.csv").drop({"Unnamed: 0"}, axis=1)

diamonds = diamonds.sample(frac=1, random_state=2023).reset_index(drop=True)

diamonds.head()
```

```python
sns.pairplot(diamonds.sample(1000), corner=True);
```

```python
cols = [
    "carat", 
#     "carat_sq", 
#     "depth", 
#     "table", 
#     "x", 
#     "y", 
#     "z"
]

X = sm.add_constant(diamonds.loc[:, cols])
y = diamonds["price"]
# y = np.log(diamonds["price"])

model = sm.OLS(y, X).fit()

model.summary()
```

```python
# residual Analysis Plotting function - residual plot and QQplot

def residual_analysis_plots(model):
    
    import scipy.stats as stats
    import matplotlib.pyplot as plt
    
    predictions = model.predict()
    residuals = model.resid
    
    fig, ax = plt.subplots(1, 2, sharey="all", figsize=(10, 6))
    
    sns.scatterplot(x=predictions, y=residuals, ax=ax[0])
    ax[0].set_title("Residual Plot")
    ax[0].set_xlabel("Prediction")
    ax[0].set_ylabel("Residuals")
    
    stats.probplot(residuals, dist="norm", plot=ax[1])
    ax[1].set_title("Normal Q-Q Plot")   
```

```python
residual_analysis_plots(model)
```

### Log Transform Target

```python
diamonds = diamonds.assign(carat_sq = diamonds["carat"] ** 2)


cols = [
    "carat", 
    "carat_sq", 
#     "depth", 
#     "table", 
#     "x", 
#     "y", 
#     "z"
]

X = sm.add_constant(diamonds.loc[:, cols])
# y = diamonds["price"]
y = np.log(diamonds["price"])

model = sm.OLS(y, X).fit()

model.summary()
```

### MultiCollinearity

```python
# Perfect Multicollinearity

import numpy as np

# Create feature x 2 column = perfectly correlated
diamonds = diamonds.assign(caratX2 = diamonds["carat"] * 2)

X = sm.add_constant(diamonds.loc[:, ["carat", "caratX2"]])
y = diamonds["price"]
# y = np.log(diamonds["price"])

model = sm.OLS(y, X).fit()

model.summary()
```

```python
diamonds[['carat', "caratX2"]].corr()
```

```python
model.params
```

```python
diamonds[["price", "carat", "x", "y", "z", "depth"]].corr()
```

### Collinearity & VIF

```python
X = sm.add_constant(diamonds.loc[:, ["carat", "x", "y", "z", "depth"]])
y = diamonds["price"]
# y = np.log(diamonds["price"])

model = sm.OLS(y, X).fit()

model.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif

pd.Series(
    [vif(X.values, i) for i in range(X.shape[1])],
    index=X.columns
)
```
